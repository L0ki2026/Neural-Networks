```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
#Generate a set of random data points using normal distrubution

np.random.seed(15)

X, Y = make_regression(n_samples=1000, n_features=1, noise=10)
X=X.flatten()
Y=Y.flatten()
plt.scatter(X, Y)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('Randomized Data points')
plt.show()
```

```python
# Building the model
```

```python
m = 0
c = 0

L = 0.0001  # The learning Rate
epochs = 10000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

E_abs = []
E_sq = []
slope=[]
Dm=[]
Dc=[]
```

```python
# Performing Gradient Descent
for i in range(epochs):
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    E_abs.append(abs((1/n) * sum((Y - (m*X + c)))))
    E_sq.append((1/n) * sum((Y - (m*X + c))**2))
    slope.append(m)
    Dm.append(D_m)
    Dc.append(D_c)
#    print(E)
```

```python
#Plot the Mean absolute error and Mean squared error for comparison
print('Loss functions')

plt.plot(E_abs)
plt.ylabel('error')
plt.xlabel('# of iterations')
plt.title('Mean absolute error')
plt.show()

plt.plot(E_sq)
plt.ylabel('error')
plt.xlabel('# of iterations')
plt.title('Mean squared error')
plt.show()

print(min(E_sq))
```
