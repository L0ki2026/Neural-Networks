```python
# Define the label_tokens function
def label_tokens(my_text):
    tokenized_words = my_text.split()  # Use word_tokenize for better tokenization
    labels = []

    # Label each token with the next token in the sentence
    for i in range(0, len(tokenized_words) - 1):
        labels.append(tokenized_words[i + 1])
    labels.append('null')

    labeled_words = list(zip(tokenized_words, labels))
    return labeled_words

example_id = gutenberg.fileids()[4]
text = gutenberg.raw(example_id)

sentences = sent_tokenize(text)

# print(sentences)

# Process each sentence with label_tokens and combine the results
labeled_text = []

for sentence in sentences:
    labeled_sentence = label_tokens(sentence)
    labeled_text.extend(labeled_sentence)

print(labeled_text[:100])
# for labeled_token in labeled_text:
#     print(labeled_token)
```

```python
word_list = []
for words, _ in labeled_text:
  word_list.append(words)
print(word_list)

label_list = []
for _, labels in labeled_text:
  label_list.append(labels)
print(label_list)
```

```python
word_pairs = [(word, idx) for idx, word in enumerate(word_list)]
vocab_map = dict(word_pairs)

print(vocab_map.get("[Poems"))
```

```python
import numpy as np

def one_hot_vector(word):
    vector = np.zeros(len(vocab_map))
    index = vocab_map.get(word)
    if index != None:
      vector[index] = 1
    return vector
print(one_hot_vector("[Poems"))
print(one_hot_vector("Mansi"))
```

```python
def get_vector_list(label_words_list):
  vector_list = []
  for word in range(0, len(label_words_list)):
    vector = one_hot_vector(word)
    vector_list.append(vector)
  return vector_list

words_vector = get_vector_list(word_list)
labels_vector = get_vector_list(label_list)

print(words_vector)
print(labels_vector)
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader, random_split

words_tensor = torch.tensor(np.array(words_vector)).float()
labels_tensor = torch.tensor(np.array(labels_vector)).float()
print(f"size(words_tensor):{words_tensor.size()}")
print(f"size(labels_tensor):{words_tensor.size()}")


dataset = TensorDataset(words_tensor, labels_tensor)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

trainset, valset = random_split(dataset, [train_size, val_size])

print(trainset)
print(valset)

trainloader = DataLoader(trainset, batch_size=64, shuffle=True)
valloader = DataLoader(valset, batch_size=64, shuffle=True)
```

```python
import torch
from torch import nn, optim
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, output_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Model, loss, and optimizer
input_size = len(words_vector[0])
output_size = len(words_vector[0])
print(f"input_size:{input_size}")
print(f"output_size:{output_size}")
model = SimpleNN(input_size, output_size)
criterion = nn.MSELoss()  # Mean Squared Error loss function for regression
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training example
words, labels = next(iter(trainloader))
print(f"Words shape: {words.shape}")
print(f"Labels shape: {labels.shape}")

# Forward pass
# outputs = model(words)  # Model outputs
# loss = criterion(outputs, labels)  # Calculate the MSE loss
# print(f"Loss: {loss.item()}")
```

```python
def one_hot_ce_loss(outputs, targets):
    criterion = nn.CrossEntropyLoss()
    _, labels = torch.max(targets, dim=1)
    return criterion(outputs, labels)

example_targets = torch.tensor([[0, 0, 1], [1, 0, 0], [0, 0, 1], [0, 1, 0]], dtype=torch.int32)
example_outputs = torch.rand(size=(4, 3), dtype=torch.float32)
example_loss = one_hot_ce_loss(example_outputs, example_targets)
print(example_loss)
```

```python
num_epochs = 10

# Training loop
for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    running_loss = 0.0

    for batch_words, batch_labels in trainloader:
        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(batch_words)
        print(outputs.shape)

        print(f"batch labels: {batch_labels.shape}")

        # Compute loss
        #loss = criterion(outputs, batch_labels)
        loss = one_hot_ce_loss(outputs, batch_labels)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        # Accumulate loss
        running_loss += loss.item() * batch_words.size(0)

    # Print average loss for this epoch
    epoch_loss = running_loss / len(trainloader.dataset)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")
    # print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss:.4f}")

    # Validation step (optional, for checking model performance on validation set)
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for batch_words, batch_labels in valloader:
            outputs = model(batch_words)
            loss = criterion(outputs, batch_labels)
            val_loss += loss.item() * batch_words.size(0)
    val_loss /= len(valloader.dataset)
    print(f"Validation Loss: {val_loss:.4f}")
```
